## Missing bmi values

```python
data = torch.load('/data/matthewc/Senior_Design/Full_Dataset.pth', weights_only=True)
count = 0
missing_bmi_count = 0
missing_bmi_idx = []
total_bmi020=0
total_bmi2040=0
total_bmi4060=0
total_bmi6080=0
total_bmi80100=0
count0_20 = 0
count20_40 = 0
count40_60 = 0
count60_80 = 0
count80_100 = 0

total = 0
for value in data:
    count += 1
    total += value[4]
    if (value[4] == 0.0): #Check if the bmi value is missing (set to 0.0)
        missing_bmi_count += 1
        missing_bmi_idx.append(count-1)
    else:    
        if(value[0] >=0 and value[0]<20):
            total_bmi020+=value[4]
            count0_20 += 1
        if(value[0] >=20 and value[0]<40):
            total_bmi2040+=value[4]
            count20_40+=1
        if(value[0] >=40 and value[0]<60):
            total_bmi4060+=value[4]
            count40_60+=1
        if(value[0] >=60 and value[0]<80):
            total_bmi6080+=value[4]
            count60_80+=1
        if(value[0] >=80 and value[0]<100):
            total_bmi80100+=value[4]
            count80_100+=1
        
print("Count = ", count)
print("Missing BMI count = ",missing_bmi_count)
print("Average BMI = ", total/(count-missing_bmi_count))
```

```python
print(missing_bmi_idx)
#These are the datapoints that are missing bmi values
```

```python
#Use Age-group mean Imputation to fill in the missing values
group0_20 = []
group20_40 = []
group40_60 = []
group60_80 = []
group80_100 = []

data = torch.load('/data/matthewc/Senior_Design/Full_Dataset.pth', weights_only=True)
for index in missing_bmi_idx:
    if(data[index,0]>=0 and data[index,0]<20):
        group0_20.append(index)
    if(data[index,0]>=20 and data[index,0]<40):
        group20_40.append(index)
    if(data[index,0]>=40 and data[index,0]<60):
        group40_60.append(index)
    if(data[index,0]>=60 and data[index,0]<80):
        group60_80.append(index)
    if(data[index,0]>=80 and data[index,0]<100):
        group80_100.append(index)
```

## Train/Test/Validation data split with SMOTE oversampling

```python
# Import statements
import torch
from imblearn.over_sampling import SMOTE

from torch import nn
from torch.utils.data import DataLoader, Dataset, random_split
import math
import time
```

```python
class StrokeDataset(torch.utils.data.Dataset):
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels
        
    def __getitem__(self,idx):
        X = self.data[idx,:]
        Y = self.labels[idx,]
        return X, Y

    def __len__(self):
        return self.data.shape[0]   
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#Set load the data and labels
data = torch.load('/data/matthewc/Senior_Design/Imputed_Dataset.pth', weights_only=True)
labels = torch.load('/data/matthewc/Senior_Design/Labels.pth', weights_only=True)
```

```python
sm = SMOTE(random_state=42) #Start tthe SMOTE implementation
resampled_data, resampled_labels =sm.fit_resample(data, labels) #Resample the data with SMOTE
# Convert to pytorch tensors
resampled_data_tensor = torch.from_numpy(resampled_data)
resampled_labels_tensor = torch.from_numpy(resampled_labels)
```

```python
Full_Dataset = StrokeDataset(data = resampled_data_tensor, labels = resampled_labels_tensor.unsqueeze(1))
#Split the full dataset into seperate training/testing datasets
train_size = Full_Dataset.__len__() *0.7
test_size = Full_Dataset.__len__() *0.15
val_size = Full_Dataset.__len__() *0.15
train_dataset, test_dataset, validation_dataset = random_split(
    Full_Dataset, [int(math.ceil(train_size)), int(math.floor(test_size)), int(math.floor(val_size))],
    generator=torch.Generator().manual_seed(42)
)
```

```python
print("Full resampled dataset size = ",Full_Dataset.__len__())
print("Training Dataset Size = ", int(math.ceil(train_size)))
print("Testing Dataset Size = ",int(math.floor(test_size)))
print("Validation Dataset Size = ",int(math.floor(val_size)))
```

# AI model

```python
class Classifier(nn.Module):
    def __init__(self):
        super(Classifier, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(21, 40), #21 input nodes
            nn.ReLU(),
            nn.Linear(40, 25),
            nn.ReLU(),
            nn.Linear(25, 10),
            nn.ReLU(),
            nn.Linear(10, 2), #2 output nodes for 2 classes
            nn.ReLU()
        )

    def forward(self, x):
        return self.net(x)
```

```python
#model = Classifier().to(device)
model = torch.load('/data/matthewc/Senior_Design/ReLU_Model.pth', weights_only = False)
#Make dataloders for the datasets
Training_Loader = DataLoader(train_dataset, batch_size = 1, shuffle = True)
Testing_Loader = DataLoader(test_dataset, batch_size = 1, shuffle = True)
Validation_Loader = DataLoader(validation_dataset, batch_size = 1, shuffle = True)

criterion = nn.CrossEntropyLoss()  # Cross Entropy Loss
optimizer = torch.optim.Adam(model.parameters(), lr=0.00005)
```

```python
epochs = 100
for epoch in range(epochs):
    model.train()
    total_loss = 0
    step = 0
    for ins, labs in Training_Loader: #For inputs/labels
        ins = ins.to(device)
        labs = labs.to(device)
        optimizer.zero_grad()
        output = model(ins)  
        loss = criterion(output, labs.squeeze(dim=0).long())
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for ins, labs in Testing_Loader:
            ins, labs = ins.to(device), labs.float().to(device)
            output = model(ins)
            predicted = output.argmax(dim=1)    
            total += 1
            correct += (predicted == labs).float()
    epoch_loss = total_loss / len(Training_Loader)
    epoch_acc = correct/(total) * 100
    print(f'Epoch {epoch+1}/{epochs}, Loss: {epoch_loss:.4f}, Test Accuracy: {epoch_acc.sum():.2f}%')
print("Done")
```

I started with 200 epoch for each different model and found that the ReLU model was the most promising. 
The 1st and 2nd models started to plateau by 200th epoch, while the 4th model plateaued around 130 epochs in. 
The 3rd model, with only ReLU for the activation function, continued to show steady improvement, so I continued  training it.   
At 1000 epochs it started to plateau aroung 94% accuracy, so I am lowering the learning rate to se if it can improve anymore.
